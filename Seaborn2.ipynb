```python

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
tips = sns.load_dataset("tips")
```

```python
tips.head(20)
```

```python
tips.shape
```

```python
tips.describe()
```

```python
# Scatter Plot: Relationship between total bill and tip
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

```python
tips.columns
```

```python
# Line Plot: Total bill over time
sns.lineplot(x="day", y="total_bill", data=tips)
plt.title("Line Plot: Total Bill over Time")
plt.show()
```

```python
#highest earning day is sunday and lowest customer visit in friday 
```

```python
# Bar Plot Avg total bill for each day
sns.barplot(x="day", y="total_bill", data=tips, ci=None)
plt.title("Bar Plot: Average Total Bill for Each Day")
plt.show()
```

```python
# Count Plot: Count of meals for each day
sns.countplot(x="day", data=tips)
plt.title("Count Plot: Count of Meals for Each Day")
plt.show()
```

```python
# Box Plot: Distribution of total bill for each day
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot: Distribution of Total Bill for Each Day")
plt.show()
```

```python

# Violin Plot: Distribution of total bill and tip for each day
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot: Distribution of Total Bill for Each Day")
plt.show()
```

```python
# Heatmap: Correlation matrix of numerical columns
sns.heatmap(tips.corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap: Correlation Matrix")
plt.show()
```

```python
# Pair Plot: Pairwise relationships in the dataset
sns.pairplot(tips)
plt.suptitle("Pair Plot: Pairwise Relationships", y=1.02)
plt.show()
```

```python
# Histogram: Distribution of total bill
sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.title("Histogram: Distribution of Total Bill")
plt.show()
```

```python
# KDE Plot: Kernel Density Estimate for total bill
sns.kdeplot(tips["total_bill"])
plt.title("KDE Plot: Kernel Density Estimate for Total Bill")
plt.show()
```

```python
# FacetGrid: Scatter plot for total bill and tip with facet grid by day
g = sns.FacetGrid(tips, col="day")
g.map(sns.scatterplot, "total_bill", "tip")
g.set_axis_labels("Total Bill", "Tip")
plt.suptitle("FacetGrid: Scatter Plot by Day", y=1.02)
plt.show()
```

```python

# Set the Seaborn theme to "ticks"
sns.set_theme(style="ticks")

# Create a scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Show the plot
plt.show()
```

```python
# Set the Seaborn theme with custom ticks
sns.set_theme(style="whitegrid", palette="pastel")
# Other themes: "darkgrid", "white", "dark", "ticks"

# Create a scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Set titles and labels
plt.title("Scatter Plot: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

# Show the plot
plt.show()
```
